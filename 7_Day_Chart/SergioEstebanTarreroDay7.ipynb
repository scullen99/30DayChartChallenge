```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fruit_prices_data = pd.read_csv('Fruit Prices 2020.csv')
fruit_prices_data.shape
```

```python
fresh_fruits_data = fruit_prices_data[fruit_prices_data['Form'] == 'Fresh']

fresh_fruits_data_sorted = fresh_fruits_data.sort_values('RetailPrice', ascending=False)

top_5_fruits = fresh_fruits_data_sorted.head(5)
```

```python
fruit_colors = {
    'Raspberries': '#FF0000',  # Raspberries are typically red
    'Blackberries': '#2E2E2E',  # Blackberries are dark blue or black
    'Strawberries': 'red',  # Strawberries are red
    'Cherries': 'red',  # Cherries are also red
    'Blueberries': 'blue',  # Blueberries are blue
    'Apricots': 'orange'  # apricots are orange
}
```

```python
bar_colors = [fruit_colors.get(fruit, '#C0C0F2') for fruit in fresh_fruits_data_sorted['Fruit']]
```

```python
plt.figure(figsize=(10, 8))
plt.barh(fresh_fruits_data_sorted['Fruit'], fresh_fruits_data_sorted['RetailPrice'], color=bar_colors)
plt.xlabel('Retail Price\n(USD per unit)')
plt.title('Retail Prices of Fresh Fruits in 2020 in the USA')
plt.gca().invert_yaxis()
plt.grid(True, which='both', axis='x', linestyle='-', linewidth=0.2, color='gray')

plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.barh(fresh_fruits_data_sorted['Fruit'], fresh_fruits_data_sorted['RetailPrice'], color=bar_colors)
plt.xlabel('Retail Price\n(USD per unit)')
plt.title('Retail Prices of Fresh Fruits in 2020 in the USA')
plt.gca().invert_yaxis() 
plt.grid(True, which='both', axis='x', linestyle='-', linewidth=0.2, color='gray')

for index, fruit in enumerate(fresh_fruits_data_sorted['Fruit']):
    price = fresh_fruits_data_sorted['RetailPrice'].iloc[index]
    plt.text(price, index, f' {price:.2f}$', va='center', fontsize=8, color='black')

plt.show()
```
