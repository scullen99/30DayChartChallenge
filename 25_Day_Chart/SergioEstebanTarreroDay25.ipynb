```python
import pandas as pd
```

```python
df = pd.read_csv('sea_levels_2015.csv')
df.shape
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['Time'] = pd.to_datetime(df['Time'])
```

```python
df['month'] = df['Time'].dt.month
df = df[df['month'] == 1]
df = df.drop('month', axis=1)
```

```python
df
```

```python
df['Time'] = df['Time'].dt.year // 5 * 5
df = df.groupby('Time').mean()
df = df.reset_index()
df
```

```python
df.to_csv('sea_levels_january_1880_to_2013_every_5_year.csv', index=False)
```
